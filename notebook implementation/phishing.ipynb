```python
MAX_UNIQUE_CATEGORIES = 30
ROWS = 6
COLS = 3
SAMPLE_SIZE = 1000

if data.shape[0] > SAMPLE_SIZE:
    data_to_plot = data.sample(n=SAMPLE_SIZE, random_state=42)
else:
    data_to_plot = data.copy()

columns_to_plot = [
    col for col in data_to_plot.columns
    if col != 'Result' and data_to_plot[col].nunique() <= MAX_UNIQUE_CATEGORIES
]

num_plots = len(columns_to_plot)
total_rows = (num_plots + COLS - 1) // COLS

fig, axes = plot.subplots(total_rows, COLS, figsize=(COLS * 5, total_rows * 4))
axes = axes.flatten()

for i, col in enumerate(columns_to_plot):
    sns.countplot(x=data_to_plot[col], ax=axes[i])
    axes[i].set_title(f'Countplot: {col}', fontsize=10)
    axes[i].set_xlabel(col, fontsize=9)
    axes[i].tick_params(axis='x', rotation=45)

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plot.tight_layout()
plot.show()
```
